### 1. Obtener una clave API de OpenRouter y guardarla en Colab

Primero, necesitar√°s una clave API de OpenRouter. Puedes obtenerla creando una cuenta en [OpenRouter](https://openrouter.ai/) y yendo a su secci√≥n de configuraci√≥n de API.

Una vez que tengas tu clave, gu√°rdala de forma segura en los secretos de Colab. Puedes acceder a los secretos haciendo clic en el icono de la "üîë" en el panel izquierdo de Colab. Dale el nombre `OPENROUTER_API_KEY`.

```python
# Instalar la biblioteca necesaria
!pip install langchain-openai --quiet
```

```python
from langchain_openai import ChatOpenAI
from IPython.display import display, Markdown
from google.colab import userdata
import os

# Obt√©n la clave API de OpenRouter de los secretos de Colab
os.environ["OPENROUTER_API_KEY"] = userdata.get('OPENROUTER_API_KEY')

# Define el base_url para OpenRouter
OPENROUTER_API_BASE = "https://openrouter.ai/api/v1"

print("Configuraci√≥n de OpenRouter lista.")
```

### 2. Inicializar el modelo a trav√©s de OpenRouter

Ahora, inicializaremos el modelo. **Importante:** Deber√°s elegir un modelo disponible en OpenRouter. Puedes ver la lista de modelos y sus proveedores (incluyendo NVIDIA si est√°n disponibles) en la p√°gina de OpenRouter. Para este ejemplo, usar√© un modelo gratuito popular, pero puedes cambiarlo por cualquier otro modelo que OpenRouter soporte y al que tengas acceso (revisa su documentaci√≥n para ver los modelos de NVIDIA que puedan ofrecer).

Algunos ejemplos de modelos de OpenRouter podr√≠an ser:
*   `mistralai/mistral-7b-instruct:free` (gratuito)
*   `google/gemma-7b-it` (gratuito)
*   `meta-llama/llama-3-8b-instruct` (si tienes acceso y es gratuito en OpenRouter)
*   Busca modelos espec√≠ficos de NVIDIA en su lista, si te interesa uno en particular (por ejemplo, un modelo Nemotron Nano si estuviera alojado all√≠).

```python
# Inicializa el modelo ChatOpenAI apuntando a OpenRouter
# Cambia 'mistralai/mistral-7b-instruct:free' por el modelo de tu elecci√≥n en OpenRouter
# Aseg√∫rate de que el modelo que elijas sea gratuito o est√© dentro de tu presupuesto de OpenRouter.
chat_model = ChatOpenAI(
    model_name="mistralai/mistral-7b-instruct:free", # <<-- Cambiado a mistralai/mistral-7b-instruct:free de nuevo
    openai_api_base=OPENROUTER_API_BASE,
    openai_api_key=os.environ["OPENROUTER_API_KEY"],
    temperature=0.7 # Puedes ajustar la temperatura para la creatividad de la respuesta
)

print(f"Modelo '{chat_model.model_name}' inicializado con OpenRouter.")
```

### 3. Generar una respuesta con el modelo de OpenRouter

Ahora, puedes usar el modelo inicializado para interactuar con √©l, de manera similar a como lo har√≠as con otros LLMs.

```python
messages = [
    {"role": "system", "content": "Eres un profesor de lenguaje de programaci√≥n en python y quieres ense√±ar"},
    {"role": "user", "content": "¬øQu√© es la herencia en Python?"},
]

# Formatea los mensajes para langchain
from langchain_core.messages import SystemMessage, HumanMessage
formatted_messages = [
    SystemMessage(content=messages[0]["content"]),
    HumanMessage(content=messages[1]["content"])
]

# Invoca el modelo para obtener una respuesta
response = chat_model.invoke(formatted_messages)

# Muestra la respuesta
display(Markdown(response.content))
```
