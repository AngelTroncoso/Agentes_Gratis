```python
# Paso 2: Inicialización (con la importación de torch y la configuración de GPU)
import torch
from langchain_huggingface.llms import HuggingFacePipeline
from transformers import BitsAndBytesConfig

# Configuración de cuantización 4-bit (sin cambios)
nf4_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Inicializamos el modelo Mistral base (compatible con BitsAndBytes)
hf = HuggingFacePipeline.from_model_id(
    # ¡Cambiamos a la versión base de Mistral!
    model_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    model_kwargs={
        "quantization_config": nf4_config,
        "device_map": "auto",
        "trust_remote_code": True # Mantenemos esto por seguridad
    },
    pipeline_kwargs={"max_new_tokens": 100},
)
```

```python
# Usamos el formato de instrucción de Mistral: [INST] Instrucción [/INST]
question_espanol = "[INST] Responde esta pregunta en español. ¿Qué es un gato? [/INST]"

hf.invoke(question_espanol)
```
